```python
import pandas as pd
df = pd.read_csv("data.csv")
df
```

```python
# calculer le prix finale(le prix final - la taxe)
```

```python
df["tax"].fillna(0, inplace=True) # remplacer les valeurs nuls pas 0
df
```

```python
df.dtypes # on le fait pour verifier le type de price car nous avons besoin d'un nombre pour faire le calcul
```

```python
# on modifie la colonne price grace a la methode apply
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))
df.price_paid = df.price_paid.astype(float)
df
```

```python
# multiplier le prix par la taxe(ildoit nous retourner 80% de la valeur)
df["price_total"] = df["price_paid"] * (1 - df["tax"] / 100)
df
```
