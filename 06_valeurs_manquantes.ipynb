```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df.isnull() # pour voir les valeurs manquantes
```

```python
df.notnull()
```

```python
df[df["tax"].isnull()]
```

```python
df[df["tax"].notnull()] # afficher les valeurs qui ne sont pas nuls
```

```python
# REMPLACER les valeurs manquantes
df["tax"].fillna(0)
```

```python
# 2eme methode 
df["tax"].fillna(method='bfill') # elle n'est pas forcement adaptée car elle utilise des données environnantes qui peut fausser l'analyse
```

```python
# enlever les rangées du tableau
df.dropna(subset=["tax"]) # cette methode est agressif car elle va supprimer toutes les rangées qui contiennent des valeurs manquantes
```
